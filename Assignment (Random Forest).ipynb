#Fraud dataset

```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import  RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data=pd.read_csv('Fraud_check.csv')
data
```

```python
data['Taxable.Income'] = data['Taxable.Income'].apply(lambda x: 'Risky' if x <= 30000 else 'Good')
data
```

```python
data.info()
```

```python
selected_coulmn=['Undergrad','Marital.Status','City.Population','Work.Experience','Urban']
X=data[selected_coulmn]
X
```

```python
Y=data['Taxable.Income']
Y
```

```python
X = pd.get_dummies(X, columns=['Undergrad', 'Marital.Status', 'Urban'], drop_first=True)
X
```

```python
x_Train, x_Test, y_Train, y_Test = train_test_split(X, Y, test_size = 0.4, random_state = 42)
```

```python
num_trees = 100

model = RandomForestClassifier(n_estimators = num_trees,
                               max_features = 3)
```

```python
model.fit(x_Train,y_Train)
```

```python
pred=model.predict(x_Test)
pred
```

```python
pd.DataFrame({'actual' : y_Test,
             'predicted' : pred})
```

```python
np.mean(pred == y_Test)
```

```python
kfold = KFold(n_splits = 10)

results = cross_val_score(model, X, Y, cv = kfold)
print(results.mean())
```

#Company dataset

```python
Data=pd.read_csv('Company_data.csv')
Data
```

```python
Data.info()
```

```python
Data=pd.get_dummies(Data, columns=['ShelveLoc','Urban','US'])
Data
```

```python
Data['Sales'] = pd.cut(Data['Sales'], bins=3, labels=['Low', 'Medium', 'High'])
Data
```

```python
Data.info()
```

```python
x=Data.iloc[:,1:]
y=Data['Sales']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
num_trees = 100

model1 = RandomForestClassifier(n_estimators = num_trees,
                               max_features = 3)
```

```python
model1.fit(x_train,y_train)
```

```python
preds=model1.predict(x_test)
preds
```

```python
np.mean(preds == y_test)
```

```python
kfold = KFold(n_splits = 10)

results = cross_val_score(model1, x, y, cv = kfold)
print(results.mean())
```
